```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import LabelEncoder

data_path = 'data/'
train_data = pd.read_csv(data_path + 'train_data.csv')
test_data = pd.read_csv(data_path + 'test_features.csv')

target_col = train_data['poi']

train_data.head()
```

```python
print('train_data: ', train_data.shape)
print('test_data: ', test_data.shape)
print('target_col: ', target_col.shape)
```

```python
train_data.describe()
```

```python
mlp = MLPRegressor(hidden_layer_sizes = (50, 60, 50), 
                   activation = 'relu', 
                   solver = 'adam', 
                   alpha = '0.0001', 
                   batch_size = 'auto',
                  learning_rate_init = 0.001,
                  power_t = 0.5,
                  max_iter = 200,
                  tol = 1e-4)
mlp.fit(train_data, target_col)

y_pred = mlp.predict(test_data)
```
