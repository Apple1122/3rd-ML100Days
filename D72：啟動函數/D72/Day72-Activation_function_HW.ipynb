# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

x = np.arange(-10, 10, 0.01)
```

# Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

plt.plot(x, sigmoid(x))
plt.grid(color='b', linewidth='0.2', linestyle='--')
plt.title('Sigmoid Function')
```

```python
def dsigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

plt.plot(x, dsigmoid(x))
plt.grid(color='r', linewidth='0.2', linestyle='--')
plt.title('Derivative Sigmoid Function')
```

# Tanh

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.plot(x, tanh(x))
plt.grid(color='b', linewidth='0.2', linestyle='--')
plt.title('Tanh Function')
```

```python
def dtanh(x):
    return 1 - np.square(tanh(x))

plt.plot(x, dtanh(x))
plt.grid(color='r', linewidth='0.2', linestyle='--')
plt.title('Derivative Tanh Function')
```

# Softmax

```python
def softmax(x):
    return np.exp(x) / float(sum(np.exp(x)))

plt.plot(x, softmax(x))
plt.grid(color='b', linewidth='0.2', linestyle='--')
plt.title('Softmax Function')
```

# ReLU

```python
def relu(x):
    return list(map(lambda x:x if x > 0 else 0, x))

plt.plot(x, relu(x))
plt.grid(color='b', linewidth='0.2', linestyle='--')
plt.title('ReLU Function')
```

```python
def drelu(x):
    return list(map(lambda x: 1 if x > 0 else 0, x))

plt.plot(x, drelu(x))
plt.grid(color='r', linewidth='0.2', linestyle='--')
plt.title('Derivative ReLU Function')
```

# Leaky ReLU

```python
def leaky_relu(x):
    return list(map(lambda x: x if x > 0 else 0.1 * x, x))

plt.plot(x, leaky_relu(x))
plt.grid(color='b', linewidth='0.2', linestyle='--')
plt.title('Leaky ReLU Function')
```
